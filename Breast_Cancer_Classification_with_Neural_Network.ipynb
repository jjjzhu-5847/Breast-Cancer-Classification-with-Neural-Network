# **Import necerry Library**

```python
import numpy as np # provide good array features
import pandas as pd # load data frame
import matplotlib.pyplot as plt # data virtualation
import sklearn.datasets # load dataset
from sklearn.preprocessing import StandardScaler # standardize data
from sklearn.model_selection import train_test_split # split data set
```

# **Load and Modify Data**

```python
# load the data
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()

# check the data correctly loaded
print(breast_cancer_dataset)
```

```python
# put data into data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)

# print first 5 rows of data frame
data_frame.head()
```

```python
# add label column to data_frame with target array in dataset
data_frame['label'] = breast_cancer_dataset.target

# print last 5 rows of data frame
data_frame.tail()
```

```python
# size of data_frame
print(data_frame.shape)

# info of data_frame
print(data_frame.info())
```

```python
# double check the non-null
data_frame.isnull().sum()
```

```python
# statistical measures about the data
# mean, median, std, min, quantile...
data_frame.describe()
```

```python
# check the distribution of label
data_frame['label'].value_counts()
```

1 --> Benign

0 --> Malignant

The reason we check is because to make sure we have even distribution of the label set. If we have all label 1 but non label 0, then we need to modify the label set like down or up sampling

```python
# group all the data with label for each column
# then calculate the mean of each groups
data_frame.groupby('label').mean()
```

```python
# sperate the data and the label
# delete column axis = 1
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']

print(X)
print(Y)
```

```python
# split the data into trainning data and testing data
# 20% of data will be testing data
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=2)

print('Original dataset size: '+ str(X.shape))
print('Trainning dataset size: ' + str(x_train.shape))
print('Testing dataset size:' + str(x_test.shape))
```

```python
# standarlize the data
x_train_std = StandardScaler().fit_transform(x_train)
x_test_std = StandardScaler().fit_transform(x_test)
```

# **Build Neural Network**

```python
import tensorflow as tf # provide network framework

# set random seed for generating random wight and bias
tf.random.set_seed(3)

from tensorflow import keras # construct network quickly
```

```python
# setting up the layers
# create a network model with three layers
# input -> hiden layer -> output
# input use Flatten to make a 30 row matrix to a one demantion vector
# hiden layer has 20 neurons with activation function ReLu
# ouput layer has 2 output, either 1 or 0. Use sigmoid for binary classfication
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# compile the Network
# optimizer, which is how to update weight or bias
# loss, which is loss function. becuase it is a Binary classification: sparse_categorical_crossentropy
# metrics shows the accuracy of the prediction
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
# train the data
history = model.fit(x_train_std, y_train, validation_split=0.1, epochs=10)
```

Visualizing the accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['trainning set', 'validation set'], loc='lower right')
plt.show()

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['trainning set', 'validation set'], loc='upper right')
plt.show()
```

Evaluate the model with test data set

```python
loss, accuary = model.evaluate(x_test_std, y_test)
```

# **Prediction**

```python
# take test record as our new case
# model.predict() give me the prediction probability of each label for each input
result = model.predict(x_test_std)
print(result)

# result will be [x,y]
# x stands for the prob it is 0, y stands for the prob it will be 1
```

```python
# convert the prediction prob to label
result_label = [int(np.argmax(i)) for i in result]
print(result_label)
```

```python
# compare result_label and y_test
correct_predictions = sum(1 for x, y in zip(result_label, y_test) if x == y)

print(f"Number of correctly predicted instances: {correct_predictions}")
print(f"Total instances in test set: {len(y_test)}")
```
